```python
bump_idx = np.arange(5)*(1000/100)+(1000/100)/2
bump_frequency = 1000/(50*2)#gives bump frequency given that bumps are defined as half-sines
np.sin(2*np.pi*np.linspace(0,1,1000)*bump_frequency)[[int(x) for x in bump_idx]]
```
